Problem 2: Highest Salary from Employee Data


Problem Statement: Store employee names and their salaries, and determine which employee has the highest salary.

```python
employees={
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_name=""
highest_sal=0
for name in employees:
    if employees[name]>highest_sal:
        highest_sal=employees[name]
        highest_name=name
print("Highest Salary:",highest_name,"-", highest_sal)
```

Problem 3: Find Maximum and Minimum Values

Problem Statement: Find the maximum and minimum values from a list of numbers.
Hint: Initialize both maximum and minimum with the first element of the list.
Use a loop to compare each element.



```python
list1=[45,22,89,10,66]
#assume first number is max and min
maximum=list1[0]
minimum=list1[0]
for num in list1:
    if num>maximum:
        maximum=num
    if num<minimum:
        minimum=num
print("List:",list1)
print("Max:",maximum)
print("Min:",minimum)
```

Problem 4: Count Products Above a Price Threshold

Problem Statement: Count how many products have a price greater than 1000.



```python
prices=[450,1200,899,1500,300]
count=0
for price in prices:
    if price>1000:
        count=count+1
print("Products above 1000:",count)
```

Problem 5: Calculate Attendance Percentage
Problem Statement: Calculate the attendance percentage from a list containing "P" (Present) and "A" (Absent).



```python
attendance=["P","P","A","P","P"]
p=0
for i in attendance:
    if i=="P":
        p+=1
total=len(attendance)
percentage=(p/total)*100
print("Attendance Percentage:", percentage)
```

Problem 6: Remove Duplicate Phone Numbers

Problem Statement: You are given a list of phone numbers that may contain duplicates. Write a Python program to remove duplicate phone numbers and display only the unique phone numbers.




```python
phone_nos=[9876543210, 9123456789, 9876543210,8767562781,8767562781]
unique=[] #empty list
for num in phone_nos:
    if num not in unique:
        unique.append(num)
print("Unique phone numbers:", unique)
```

Problem 7: Count Character Frequency
Problem Statement: count how many times each character appears in a given string.



```python
string="tanishkajadhav"
result={} #empty dict
for i in string:
  #If character already exists in dictionary: Increase its value by 1 ELSE If character does not exist: add 1
    if i in result:
        result[i]+=1
    else:
        result[i]=1
print(result)
```
